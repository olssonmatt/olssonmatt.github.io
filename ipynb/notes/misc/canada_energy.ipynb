# NRCAN


## Canada Greener Homes Initiative
 - __[LINK](https://natural-resources.canada.ca/energy-efficiency/homes/canada-greener-homes-initiative/24831)__
 - To help Canadians live more energy-efficient, save money, and *create jobs* for energy advisors to fight climate change
 - Canada Greener Homes Grant
     - An open application process that applies for a grant to retrofit a houses insulation, windows, doors, heat pumps, solar panels and resiliency measures
         - Resilience in renewable energy = renewable energy systems that are resistant to global shocks as per __[ren21](https://www.ren21.net/renewable-energy-resilient/#:~:text=Renewable%20energy%20systems%20are%20resistant,those%20based%20on%20fossil%20fuels.)__
    - Need to take a pre and post-retrofit EnerGuide evaluation of the home to be eligible for the grant
    - Grants range from \\$125 to \\$5,000 to get back from the retrofits
    - \\$0- \\$600 can be contriubited towards the EnerGuide evaluations
 - Canada Greener Homes Loan
    - An open application process that is an interest-free financing (loan) to complete major retrofits recommended by an energy advisor
    - Can only be combined with the above grant
    - Interest-free loans can range from \\$5,000 to \\$40,000 with a repayment term of 10 years
 - Oil to Heat Pump Affordability Program
    - An open applciation that seeks to help households with the median income or less who are heating their homes with oil to make an energy efficient change
    - NO evaluation needed
    - \\$0 to \\$10,000 to cover the costs for changing the oil heating system to a cold climate air source heat pump
    - Would include \\$0 to \\$5,000 from the Canada Greener Homes Grant
   
## How to become an energy advisor
 - __[LINK]()__
 - The following steps are to become a registered energy advisor for NRCAN and would be able to give EnerGuide Rating Systems
    1. Get Recruited: A service organization recruits you (or you them) to become an energy advisor candidate
    2. Meet Security Criteria: The service organization undertakes a pre-employment security screening
    3. Start Learning: Learn about EnerGuide program, home, and building science principles from the serivce organization to prepare for the training & examination
    4. Be trained and mentored: Trained/mentored on home/building science procedures through the initiative by the serivce organization or independent courses
    5. Pass the Exams: Take and pass the __[Foundation Level](https://natural-resources.canada.ca/energy-efficiency/homes/professional-opportunities/become-registered-energuide-rating-system-energy-advisor/20566#fle)__ and __[Energy Advisor](https://natural-resources.canada.ca/energy-efficiency/homes/professional-opportunities/become-registered-energuide-rating-system-energy-advisor/20566#eaeh)__ exams
 - Can apply to become a licensed service organization with the following critiea
    - be an individual or legal entity incorporated or registered in Canada
    - show proof of minimum \\$2$ million in liability insurance
    - have a minimum of two years experience and knowledge related to energy efficiency and building science in the residential housing sector
    - employ key personnel in the organization such as manager(s), quality assurance specialist(s), and energy advisor(s) who have successfully passed the NRCan exams

## Renewable Energy
 - __[LINK](https://natural-resources.canada.ca/our-natural-resources/energy-sources-distribution/renewable-energy/7293)__
 - Key Descriptors
     - Canada, with its large landmass and diversified geography, has substantial renewable resources that can be used to produce energy; these resources include moving water, wind, biomass, solar, geothermal, and ocean energy.
    - Canada is a world leader in renewable resources. Renewable energy sources currently provide about 18.9 per cent of Canada’s total primary energy supply.
    - Moving water is the most important renewable energy source in Canada, providing 59.3 per cent of Canada’s electricity generation
    - Wind is the second most important renewable energy source in Canada. It accounts for 3.5 per cent of electricity generation in Canada.
    - Biomass is the third largest renewable source of Canada’s electricity generation. Its share in Canada’s electricity generation is 1.4 per cent.
    - ***Wind and solar photovoltaic energy are the fastest growing sources of electricity in Canada.***
 - What is renewable energy?
     - It is energy that is derived from natural processes that are replenished at a rate that is equal to or faster than the rate at which they are consumer
         - $\frac{\partial E_{replenished}}{\partial t} \geq \frac{\partial E_{consumed}}{\partial t}$
 - Solar energy
     - It is energy from the sun in the form of radiated heat and light
     - Historically done through passive solar technologies (?)
     - Currently, two active solar technologies involve electrical or mechanical or more common
         1. Solar 'collectors' or panels are used to heat water or ventilation air
         2. Solar photovoltaic techonology uses solar cells to convert sunlight into electricity
             - Some PV cells can convert artificial light into electricity
             - Wonder if this can leverage street light or what other artificial sources can be taken advantage of
     - Potential of solar in costal areas of Canada is low due to cloud coverage and higher in central regions
     - ***They estimate half of Canada's residential electricity requiremetn could be met by solar***
        - An important measurement considering roughly half of Canada's population are in southern Ontario
 - __[Solar Photovoltaic Energy](https://natural-resources.canada.ca/energy/energy-sources-distribution/renewables/solar-photovoltaic-energy/7303)__
     - Canada has favoured photovoltaic (PV) technology to be a renewable energy source due to the following factors; need to reduce greenhouse gases, deregulation?, and restructing electric power generating companies
    - Primary mandate is to help develop and deploy photocoltaic energy technologies in Canada
    - This is done by firstly by accelerating the deployement of solar power in Canada, and secondly by exploiting solar energy's potential nationally and internationally
    - CanmetENERGY projects
        - Integration of utility-scale photovoltaic applications within smart grids
        - Research and demonstration on low energy solar homes
        - Development of codes, certification, and installation standards for PV systems and components
        - Developing photovoltaic and solar resource maps for Canada and participating in international collaboration on solar PV electricity forecasting
        - Representing Canada in the International Energy Agency Photovoltaic Power Systems Programme; see the Canadian PV Annual Report
        - Disseminating information to the Canadian PV industry; see our list of publications and success stories

# CIET


## About
 - CIET =  Canadian Institute for Energy Training
 - Offers a wide range of sustainable energy training programs
 - Also offers consulting services on energy training, ranging from corporate training assessments and gap analyses to comprehensive energy training implementation strategies and curriculum development
     - __[GAP Analysis](https://en.wikipedia.org/wiki/Gap_analysis)__ = measure of actual performance to potential or desired performance. Follows the pillars of usage gap, existing gap, product gap
 - Who are we?
    - High-quality energy efficiency training center
    - Developer and provider of the exclusive Certified RETScreen® Expert (CRE) course
    - Trained more than 20,000 professionals
    - Provider of 20 standardized training programs
    - Exclusive AEE (Association of Energy Engineers) partner for training and certification programs in Canada
    - Exclusive delivery agent of National Resources Canada for the Advanced Building Recommissioning Course
 - What do we do?
    - Deliver several certification programs, including:
        - Certified Measurement and Verification Professional (CMVP)
        - Certified Energy Manager (CEM)
        - Certified Energy Auditor (CEA)
        - Certified RETScreen® Expert (CRE)
    - Development of customized energy training programs
    - Public, on-site, and virtual real-time classrooms.
    - Customized training offered anywhere.
    - Certification body for recognized training programs
    
## Programs
 - Energy management
    - Process of tracking optimizing, and reducing energy consumption in a building or organization
    - Persons would drive projects relating to energy data collection and analyzation, reduction by implementation of energy efficient manners, optmization of equipment, and incorporation of renewable energy
 - Energy Consulting
    - Helps organizations understand energy use while making it more cost-effective and efficient
    - Required to collect and analyze energy data to establish baseline usage and set targets
    - Analysis to identify which systems use the most energy and which system offer the best opportunities for savings
 - NRCan-Registered Energy Advisor Program
    - Advisors help Canadian homeowners make better choice about energy saving retrofits or upgrades
    - Use building science, and construction knowledge of the HOT2000 energy modelling softare to conduct residential energy assessments and deliver EnerGuide ratings on eligible homes
        - __[HOT2000](https://natural-resources.canada.ca/energy-efficiency/homes/professional-opportunities/tools-industry-professionals/20596)__ has a download of a form of the program

# Glossary

 - **Building Science**: A body of knowledge that draws on engineering, architecture, chemistry, physics, biology, and life sciences to understand the physical behaviour of a building as a system
 - **Energy Efficency**: the use of less energy to perform the same task or produce the same result
 - **HOT2000**: An energy simulation modelling softare developed and maintained by NRCan to support the EnerGuide Rating ststem, ENERGY STAR for new homes, and R-2000 residential energy efficiency initiatives
 - **PV**: Photovoltaic
 - **R-2000**: A standard for building that promotes the use of affordable, energy efficient building practices and techniques, clean air features and other measures to help protect the environment
 - **Renewable Energy**: Energy that is derived from natural processes that are replenished at a rate that is equal to or faster than the rate at which they are consumer
 - **Resilience**: Renewable energy systems that are resistant to global shocks
