# oscillations and waves


### net force
 - Net force determines an objects trajectory
 - | Net Force |  |
   | --- | --- |
   | Constant | Straight line if v is parallel with F, if not then parabola |
   | Centripetal | Circular (has constant magnitude but center facing) |
   | Spring Force | Produces Oscillations |
      
### spring force
 - Refer to Hooke’s Law
 - F = -kx, where x is displacement from equilibrium, and k is the spring constant
 - Spring forces produces oscillations
 
### uniform circular motion
 - Position, $\theta = \omega t$, where $\omega$ is angular velocity
 - $\omega=2\pi f$, where $f$ is frequency
 - Velocity, $v = -A\omega\sin(\omega t)$, $A =$ amplitude
 - Acceleration, $a = -A\omega^2\cos(\omega t)$
 - Important Relation
    - $m\omega^2 = k$

### energy in simple harmonic oscillators
 - We can’t use $W=Fd\cos\theta$ because it is an oversimplification for work
 - Since is the force is constantly changing we cannot accurately define it using that equation
 - $U_{Spring} = \frac{kx^2}{2}$
 - Note: Never use kinematics in SHO, $a$ is not constant

# electric fields

### concept of a field
 - A field is a way of conceptualizing and mapping a force that surrounds any object and acts on another object at a distance without any apparent physical contact
 - In Coulombs Law, the equation depends on a point charge and test charge
 - To simplify things we prefer that the field depends on the point charge rather than the test charge
 - $E = \frac{F}{q}$, where $F$ is the Coulomb force and $q$ is the point charge, $E$ is in $\frac{Newtons}{Coulomb}$
 - $E = \frac{kQ}{r^2}$

# electric field lines

### definition
 - Drawing lines to represent electric fields around charged objects are useful in visual strength and direction since it is a vector
 - For an electric charge, if it is negative they point radially inward
 - For positive charges, they point radially outward

### electric field due to ‘continuous’ charge distribution

### motion of $e^-$ in constant $E$

# electric potential and electric field


### electric potential energy: potential difference
 - When a free charge is accelerated in an electric field it gains kinetic energy
 - The sign on the charge indicates in which what it will accelerate
 - We know $W=Fd\cos\theta$ and $F=qE$ so we define electric potential as $V=\frac{U}{q}$ 
 - This is electrical potential energy per unit charge 
 - 1 electron volt $= 1 eV = 1.602\times 10^{-19}$

### electric potential in a uniform electric field
 - Parallel plates allow for simple equations to be made 
 - For parallel plates, $E = \frac{V}{d}$
 - This is the relationship of the electric field and voltage for parallel plates and uniform electric field

### electrical potential due to a point charge
 - The electric potential for a point charge is given by $V = \frac{kQ}{r}$

### equipotential lines
 - Electric fields and electric potential are heavily related
 - We can imagine this graphically by drawing lines where the electric potential is constant
 - When these two are combined, we can see that the electric field and the electric potential are perpendicular
 - The electric potential lines are called equipotential lines

### capacitors and dielectrics
 - Capacitors
    - A capacitor is a device that is used to store charge
    - The amount of charge a capacitor can store depends on two major factors which are the applied voltage and size
    - The parallel plate capacitor is the simplest form of this
    - Capacitance can be calculated by the formula $C = \frac{Q}{V}$
    - Coming back to the parallel plate capacitor, the capacitance can be found in the equation $C =\frac{\epsilon A}{d}$,  where $A$ is the area of the plates, and $\epsilon$ is the emissivity of free space
 - Dielectrics
    - You can only make the distance so small to increase capacitance
    - The space in between the capacitors can be filled with an insulator called a dielectric to increase capacitance
    - $C = \frac{k\epsilon A}{d}$, where $k$ is the dielectric constant

### capacitors in series and parallel
 - Capacitors in Series
     - $\frac{1}{C_{Tot}} = \frac{1}{C_1} + \frac{1}{C_2} …$
 - Capacitors in Parallel
    - $C_{Tot} = C_1 + C_2 …$

### energy stored in capacitors 
 - Capacitors store potential energy
 - It is stored in the electric field
 - $U = \frac{CV^2}{2} = \frac{QV}{2} = \frac{Q^2}{2C} $

# electric current, resistance, and Ohm's law

### current 
 - Current is the rate at which charge flows
 - $I = \frac{\Delta Q}{\Delta t}$
 - Drift Velocity
    - Electric signals move very fast, sometimes at $10^8 \frac{m}{s}$
    - The average velocity of the electron is $\approx 0$
    - Drift speed is introduced, which tells us the average velocity of the free charges
    - $v_d = \frac{qEt}{m}$, where $q$ is the charge, $E$ is the electric field, $t$ is the time in-between collision, and $m$ is the mass
 - Drift Velocity and current have a relationship together
 - $I = nqAv_d$ , where $n$ is the number of free charges per unit volume, $q$ is the charge…

### Ohm’s law: resistance and simple circuits
 - Ohms Law: German physicist Georg Simon Ohm demonstrated that current flows through most substances with an applied voltage
 - He found $I \propto V$
 - Resistance and Simple Circuits
    - The electric property that impedes current is called resistance
    - Resistance is, $R \propto \frac{1}{I}$
    - This gives the relationship $V = IR$
    - Resistance is in ohms ($\Omega$)

### resistance and resistivity
 - Material and shape dependence of resistance
 - Resistance is proportional to its length and is inversely proportional to its cross-sectional area
 - $R = \frac{pL}{A}$, where $p$ is the resistivity, $L$ is the length, and $A$ is the area

### electric power and energy
 - $P = IV = \frac{V^2}{R} = I^2R$

### alternating current vs. direct current
 - Direct Current is the flow of electric charge in one direction
 - Alternating Current is the flow of electric charge that periodically reverses direction
 - Average Power $= \frac{IV}{2}$

# circuits and DC instruments

### resistors in series and parallel
 - Resistors in Series	
    - $R_{Tot} = R_1 + R_2…$
 - Resistors in Parallel 
     - $\frac{1}{R_{Tot}}= \frac{1}{R_1} + \frac{1}{R_2}…$

### Kirchhoff’s rules
 - Kirchhoff’s Current Rule (KCR)
    - The sum of all the currents entering a junction must equal the sum of all currents leaving the junction
    - Conservation of charge
 - Kirchhoff’s Voltage Rule (KVR)
    - The algebraic sum of changes in potential around any closed circuit path must be zero
    - $V_0 – IR_1 – IR_2 … = 0$

### DC voltmeters and ammeters
 - Voltmeters
    - Voltmeters are connected in parallel with whatever devices voltage is being measured 
    - It is in parallel because objects in parallel achieve the same potential difference
    - Voltmeters have a very large resistance
    - $V = I(R + r)$, where $r$ is a shunt resistor
 - Ammeters
    - Ammeters are connected in series with whatever devices current is being measured
    - A series connection is used because objects in series have the same current
    - Ammeters have small resistors

### DC circuits containing resistors and capacitors
 - An RC circuit is a circuit containing both a resistor and a capacitor
 - Charging Capacitor
    - $\Delta V_C = V_0 (1 - e^{\frac{-t}{RC}})$, where $t$ is time, and $RC$ is resistance and capacitance
 - Discharging Capacitor 
    - $\Delta V_C = V_0(e^{\frac{-t}{RC}})$

# magnetism

### magnets 
 - Magnets have universal characteristics
 - Like poles repel, unlike poles attract
 - It is impossible to have separate poles (splitting magnets just smaller magnets)
 - A bar magnet is a magnetic dipole

### ferromagnets and electromagnets
 - Few materials have strong magnetic effects
 - Those materials are ferromagnetic
 - Ferromagnets can be magnetized or induced to be made magnetic
 - Electromagnetism is the use of current to make magnetic fields and it also makes temporary magnets
 - Current is the source of magnetism

### magnetic fields
 - The direction of the magnetic field is tangent to the field
 - The strength of the field is proportional to the closeness of the lines
 - Magnetic field lines can never cross
 - Magnetic field lines are continuous, no beginning or end

### magnetic field strength: force on a moving charge
 - It is important to understand that moving charge (current) creates the magnetic field but charges also experience the magnetic field
 - This can be calculated and is seen in the equation 
    - $F_{mag} = qvB\sin\theta$, where $q$ is charge, $v$ is velocity, $B$ is the magnetic field, $\theta$ is the angle between $B$ and $v$
 - The direction of the magnetic force can be found by the right-hand rule

### force on a moving charge applications
 - Magnetic Force can cause a charged particle to move in a circular path
 - This relation links magnetic force and uniform centripetal motion
 - $qvB = \frac{mv^2}{r}$

### the Hall effect
 - Magnetic field effects charges moving in a conductor
 - The voltage across a current carrying conductor by a magnetic field is known as the Hall effect
 - $V = vBI$

### magnetic force on a current carrying conductor
 - The magnetic force on charge moving in a conductor is transmitted to the conductor itself
 - $F = ILB\sin\theta$
 - Direction of the magnetic field is formed by the corkscrew rule

### magnetic fields produced by currents
 - Magnetic field created by a long straight wire
    - Same concept as corkscrew rule
    - $B = \frac{\mu I}{2R}$, where $\mu$ is $4\times 10^{-7}$, and $r$ is the distance from the wire
 - Magnetic field created by a single loop
    - $B = \frac{\mu I}{2R}$
 - Magnetic field created by a solenoid
    - $B = n\mu I$,  where $n$ is the number of turns per unit length
 - Note: Currents in parallel in the same direction repel but currents in parallel in the opposite direction attract

# electromagnetic induction, and AC circuits

### induced EMF and magnetic flux
 - Magnetic flux is a measurement of total magnetic field which passes through an area
 - $\Phi = BA\cos\theta$ , where $B$ is magnetic field, $A$ is the area, and $\theta$ is the angle between $B$ and the normal
 - The three ways to change flux would be by changing the size of $B$, changing the area, or changing the angle
 - Magnetic Flux also produces an EMF or voltage/potential difference 
 - $V = \frac{-\Delta\Phi}{\Delta t}$
 - This is called Faraday’s Law and is described as the change in flux over time

### Faraday’s law of induction: Lenz’s law 
 - $V = \frac{-N\Delta\Phi}{\Delta t}$, $N$ is the number of turns in a coil
 - The negative sign in the equation was introduced by Lenz’s Law which states the direction of the current induced by a changing magnetic field will create a field that opposes it
 - If flux is increasing, induced $B$ in opposite direction
 - If flux is decreasing, induced $B$ in same direction

### motional EMF
 - This chapter shows how a rod in a circuit can be used to calculate things like speed of the rod etc.
 - $V = -Bwv$

### electric generators
 - Electric generators turn mechanical energy into electrical energy
 - This is all based-on changes in the angle
 - $V = -NBA\omega\sin(\omega t)$

### transformers
 - Transformer do what their name implies, they transform voltages from one value to another
 - They do this by having coils to introduce a flux
 - The flux in the coils are equal as well, this allows us to write the expression
 - $\frac{V_1}{N_1}=\frac{V_2}{N_2}$
 - Also, if there is no power loss, we find the relation
    -  $\frac{V_1}{V_2}=\frac{N_1}{N_2}=\frac{I_2}{I_1}$

# electromagnetic waves

### Maxwell’s equation
 - James Clark Maxwell made a set of equations that show the relationship of EM
 - Maxwell’s Equations state
    - Electric field lines originate on positive charges and terminate on negative charges. The strength of the force is also related to the electric constant
    - Magnetic fields are continuous and magnetic monopoles do not exist. The strength of the magnetic force is related to the magnetic constant
    - A changing magnetic field induces and electric field and the direction of it opposes the change
    - Magnetic fields are generated by moving charges or by changing electric field
 - Maxwell also found that $\frac{1}{\sqrt{\mu\epsilon}}$ is equal to the speed of light

### production of electromagnetic waves
 - Electro and magnetic waves propagate perpendicular to each other
 - Electromagnetic waves are transverse waves
 - $E = cB$, where $E$ is the electric field, $c$ is the speed of light, $B$ is the magnetic field

### the electromagnetic spectrum
 - Frequency, wavelength, and speed have a relationship together
 - It is $c = \lambda f$, where $c$ is the speed of light, $\lambda$ is the wavelength, $f$ is frequency

### energy in electromagnetic waves
 - Energy Density $= \frac{Energy}{Volume} = \frac{\epsilon E^2}{2} + \frac{B^2}{2\mu}$ 
 - Intensity $= \frac{Power}{Area} = \frac{\epsilon E^2 c}{2}$

# wave optics

### Young’s double slit experiment 
 - In Young’s experiment, he displayed that light acts like a wave
    - Bright fringe: $\sin\theta = \frac{m\lambda}{d}$
    - Dark fringe: $\sin\theta = \frac{(m+\frac{1}{2})\lambda}{d}$ 

### single slit interference
 - Light also interferes with single slits
    - Dark Fringe: $a\sin\theta = m\lambda$

### thin film interference
 - Because of reflection and refraction, light waves split up when passing through a thin film
 - $\lambda_{film} = \frac{\lambda}{n}$

### diffraction
 - Diffraction is the bending of waves around obstacles or an edge of an opening
 - Huygens’ Principle
    - Every point on a wave front acts as a source of tiny wavelets that move forward with the same speed as the wave; the wave front at a later instant is the surface that is tangent to the wavelets.
 - Dark Fringes: $\sin\theta = \frac{m\lambda}{a}$
 - Circular Diffraction
    - $\sin\theta = \frac{1.22\lambda}{D} $
 - Diffraction Grating is where there is an arrangement of closely spaced parallel slits
    - Principle Maxima of Diffraction Grating: $\sin\theta = \frac{m\lambda}{d}$

### polarization
 - Polarization by Reflection
    - Happens at Brewsters Angle
    - $\theta_B = \tan^{-1}(\frac{n_2}{n_1})$
 - Polarization
    - $I_{Transmitted}=I_{\theta}\cos^2\theta$

# quantum physics


### wave-particle duality
 - Light is a stream of particles
 - Light is an EM wave
 - Both motives are proved by different experiments
 - Photons are light particles
 - $E = hf$
 - $p = \frac{h}{\lambda}$
 - Intensity $= nhf$, where $n$ is the number of photons per area time
 - Intensity $= \frac{\epsilon cE^2}{2}$

### photoelectric effect
 - The photoelectric effect is an ejection of electrons from metal when shining light on a metal
 - Stopping Potential
    - $KE_{Lost} = e\Delta V$
    - When no $e^-$ at all cross this $\Delta V$ is the ‘stopping potential’
 - Threshold Frequency
    - Smallest frequency above which photoelectrons are ejected
    - If $f < f_0$, no photoelectrons are ejected no matter how intense
### photon absorption
 - All photon energy is given to a single electron, the photon disappears and its energy becomes $KE$ of $e^-$ 
 - Conservation of Energy
    - $\frac{mv^2}{2} = hf – W_F$, where $W_F$ is the work function or binding energy

### compton scattering
 - $\Delta\lambda = \frac{h}{mc}(1 – \cos\theta)$
 - Scattered x-ray photon has less energy than the incident one because some energy is transferred to “recoil electron”

### matter waves
 - $p = \frac{h}{\lambda}$
 - $λ = \frac{h}{mv}$

### Heisenberg uncertainty principle
 - $\Delta x\Delta p \geq \frac{h}{4\pi}$, where $\Delta x$ is the uncertainty in $x$, and $\Delta p$ is the uncertainty in momentum

### quantum particle in a box
 - In QM, a particle is described by a wave function
 - This wave function is calculated by solving Schrodinger’s Equation
 - The probability of finding the particle at between position $x$ and $x + \Delta x$ 
 - Let us put a particle in a box it cannot escape
    - Allowed wavelengths of the particle are $\lambda = \frac{2L}{n}$
    - $E_{Tot} = \frac{h^2 n^2}{8 m L^2}$
 - In classical mechanics, a particle can have only kinetic energy. In QM, there are only certain energy states allowed
 - Lowest possible energy state is when $n = 1$

# atomic physics

### Bohr model
 - Postulates
    - $e^-$ can exist in special non-radiating “stationary states”
        - Momentum $= L = n\hbar$
    - e- performs uniform circular motion about the nucleus
        - $r = n^2(\frac{h^2}{4 \pi^2 k e^2 m})$

### energy in particles
 - $E = hf = 13.6 eV (\frac{1}{nf^2} - \frac{1}{ni^2} )$

### quantum mechanics of the atom
 - There are 4 quantum numbers
 - $n$ – principal states $= 1, 2, 3, 4, ...$
 - $l$ – angular momentum $= 0, 1, 2, ... n-1$
 - $m$ – z component of angular momentum $= -l, -l + 1, ..., 0, ..., l - 1, l$
 - $s$ – spin angular momentum = $\pm\frac{1}{2}$
 - Pauli Principle: Each electron in a multi-electron atom must have a different set of quantum numbers (each $e^-$ has its own state)
