# kinematics

### displacement
 - Position
    - A vector quantity
    - Is used for reference frames
 - Displacement
    - A vector quantity
    - $\Delta x = x_2 – x_1$
 - Distance
     - Scalar quantity
     - Total length travelled
     - Addition rather than the displacements subtraction

### vectors, scalars, and coordinate systems
 - Scalar is just a magnitude, vector is a magnitude and a direction
 - In one dimension, vectors can be represented by $(+)$ or $(-)$
 - A simple 4 quadrant coordinate system can be used

### time, telocity, and speed
 - Time
    - Used as an interval and measures change
    - SI unit = seconds
    - Elapsed time $= \Delta t = t_2 – t_1$
 - Velocity
    - Velocity is a vector quantity
    - Average velocity $= \frac{displacement}{time}$
 - Speed
    - Speed is a scalar quantity
    - Average speed $= \frac{distance}{time}$
 - Graphs
    - Refer to notebook for information about position-time, velocity-time, and speed-time graphs
 - Acceleration
    - Acceleration is the change in velocity in relation to time
    - Since velocity is the measurement, acceleration is a vector quantity but it can be a scalar if measured with speed
    - Average Acceleration $= \frac{v_2 - v_1}{t}$
    - When an object slows down, its acceleration is opposites to the direction of motion, this is deceleration

### motion equations for constant acceleration in one dimension
 - $v_2 = v_1 + at$, doesn’t have displacement
 - $d = v_1t + \frac{at}{2}$, doesn’t have final velocity
 - $v_2^2 = v_1^2 + 2ad$, doesn’t have time

### gravity
 - Objects in freefall experience the force of gravity fully
 - Gravity is a force that accelerates an object
 - It accelerates a rate of $9.81 \frac{m}{s^2}$ as variable $g$

### falling objects
 - Gravity
    - If air resistance and friction are negligible, then all objects fall toward Earth at the same constant acceleration, independent of mass
    - Heavier objects fall faster than lighter objects because of air resistance and friction
    - An object that falls with no air resistance and no friction is said to be in freefall
    - $g = 9.81 \frac{m}{s^2}$ 
 - One Dimensional Motion Involving Gravity
    - When an objet is dropped initial velocity is $0$
    - When an object reaches max height, $v_y= 0$

### graphical analysis of one dimensional motion
 - Slope and General Relationship
    - When 2 physical quantities are plotted on a graph the horizontal axis is considered to be independent of the vertical axis (independent and dependent variables)

# 2D kinematics

### kinematics: 2 dimensions
 - In one-dimensional kinematics, arrows were used to denote vectors
 - In two-dimensional kinematics, a magnitude can have 3 parts: horizontal vector, vertical vector, and average vector
 - The addition of the vertical and horizontal vectors gives us the average vector
 - In most cases, basic trig and Pythagorean’s theorem can be used
 - The Independence of Perpendicular Motions
    - When a person walks eastward, their position con only more when there is movement in that phone
    - Vice Versa when you walk northward, your position in that plane is only affected by movement in the southern or northern direction
    - The horizontal and vertical components of two-dimensional are independent of each other; any motion in the vertical direction doesn’t affect the horizontal and vice versa

### vector addition and subtraction: graphical methods	
 - Vectors in Two Dimensions
    - A vector is a quantity with a magnitude and a direction
    - In a one dimensional motion this could be represented by a $+$ or $–$
    - In two dimensional however, vectors will need reference frames
 - Vector Addition: Tip-to-Tail
    - This is a graphical way to add vectors
    - The tip of one vector lines up with the tail of the next vector
    - If the vectors make a right angle, use Pythagorean’s theorem and trig
 - Vector Subtraction
    - Vector Subtraction is adding the negative of a vector
    - $A - B = A + (-B)$
 - Multiplication of Vectors and Scalars
    - When a vector is multiplied by a scalar
        - The magnitude of the vector becomes the absolute value of the vector
        - If the scalar is positive, the magnitude changes and the direction stays the same
        - If the scalar is negative, the magnitude and the direction changes
 - Resolving a Vector into Components
     - When we add vectors we achieve a resultant vector
     - To get to this we reverse engineer the resultant into $x$ and $y$ component
     - We use Pythagorean’s theorem and trig to get there

### vector addition and subtraction: analytical methods
 - Analytical methods of vector addition use geometry and simple trig
 - Resolving a Vector into Perpendicular Components
     - Right angle triangles are a necessity in analytical methods
     - Refer to notebook for in depth analysis

 - Projectile Motion
    - the motion of an object or projected into the air which is only subjected to gravity

# forces and newton's laws of motion

### development of force concept
 - Dynamics- the study of forces that cause objects and systems to move
 - Force- a push or pull with a magnitude and direction
 - Free-Body Diagrams are used to display forces and other information
 - Four Fundamental Types of Forces
      - Strong Nuclear Force
      - Weak Nuclear Force
      - Electromagnetic Force 
      - Gravitational Force

### newton’s first law of motion: inertia
 - “A body at rest remains at rest, or if in motion, remains in motion at a constant velocity unless acted on by a net external force
 - Mass
      - The inertia of an object is measured by its mass
      - Mass is the measure of the amount of matter in an object
      - Mass is constant in the universe, while weight changes depending on gravity
      - Weight $= mg$
 - Reference Frames
      - Use overhead/side views when thinking about situations because our body is not a good reference frame

### newton’s second law of motion: concept of a system
 - Newton’s Second Law is closely related to his first law
 - His second is quantitative
 - The second law contains the relationship of force, mass, and acceleration
 - The net force cause a change in motion(acceleration) of a mass
 - Acceleration is directly proportional to the net force ($a ∝ F_{net}$)
 - Mass in inversely proportional to acceleration $(a ∝ \frac{1}{m})$
 - Therefore, $F_{net} = ma$
 - Force is measured in $N$ (newtons) or $kg \frac{m}{s^2}$
 - Weight is also found by multiplied the mass by gravity

### newton’s third law of motion: symmetry in forces
 - Whenever one boy exerts a force on a second body, the first experiences a force that is equal in magnitude and opposite in direction to the force that it exerts
 - $F_{A on B} = -F_{B on A}$

### normal, tension, and other examples of force
 - Normal Force
      - Weight or force of gravity is a constant force that is always present
      - Some other force must counter act it to keep objects from constantly falling
      - This is the normal force
      - When an object is stationary in the vertical direction the net force is zero (equilibrium) so the normal force is equal to the force of gravity
 - Tension
      - Tension is the force along the length of a medium
      - Any flexible connector (rope, string, cable) can exert pulls parallel to its length
      - Tension is a pull in a connector
      - Tension must equal the force of gravity or else the medium would break

### further application of newton’s laws of motion
 - Drag force
     - Drag force is a force like friction that opposes motion and is present in air and water
     - There is no set formula for drag force

### 4 basic forces
 - The fact that there are only 4 forces in physics is one of the most remarkable simplifications
 - Nearly all of the forces we experience are just the electromagnetic even

# uniform circular motion and gravitation

### rotation angle and angular velocity
 - When an object rotates about an axis it follows a circular arc
 - There is also a rotation angle that is present
 - $\Delta \theta = \frac{\Delta S}{r}$, where $\theta$ is the change in the rotation angle, $S$ is the arc length, and r is the radius
 - Since we know 1 revolution is $2\pi$ radians, we can use this as a conversion from revolution to radians
 - Angular Velocity
    - Angular velocity is the rate of change of an angle
    - $\omega = \frac{\Delta \theta}{\Delta t}$, where $\theta$ is the angular rotation, $t$ is the time, and $\omega$ is the angular velocity and the units are $\frac{rad}{s}$
    - Angular velocity is closely related to linear velocity
    - $v = rω$

### centripetal acceleration
 - When an object moves in a circular path its direction is constantly changing
 - Since the direction if changing it has centripetal acceleration ($a_c$)
 - Centripetal acceleration moves around an axis and its direction always faces the axis of rotation
 - $a_c = \frac{v^2}{r}$, where $a$ is the centripetal acceleration, $v$ is the velocity, and $r$ is the radius
 - $a_c$ can also be calculated by the formula $a_c$ = $r\omega^2$ 

### centripetal force
 - Any net force dealing with uniform circular motion is called a centripetal force
 - $F_c = ma_c$
 - Since we have formulas for centripetal acceleration, those can be subbed into the centripetal force equation
 - It should be noted that when friction is applied, the force of friction points toward the centre as well
 - Banked Curves
    - For ideal banking, the net force is equal to the horizontal centripetal force
    - These cases are similar to inclined planes
    - Instead the normal force is broken up into components rather than tipping the coordinate system

### Newton’s universal law of gravitation
 - Gravity is a simple force, it always attracts a mass and the space between them
 - Newton’s Universal Law of Gravitation states that every particle in the universe attracts every other particle with a force joining them
 - $Fg = \frac{Gm_1m_2}{r^2}$ , where $G$ is the gravitational constant, $m_1$ and $m_2$ are the masses which are being analyzed and r is the distance between the centre of masses
 - The gravitational constant is $6.674\times10^{-11}$

### satellites and Kepler’s laws: an argument for simplicity
 - Orbits have 2 very important characteristics	
    1. A small mass, $m$, orbits a much larger mass, $M$. this allows us to view the motion as if $M$ were stationary- in fact, as if from an inertial frame of reference were placed on $M$. Mass $m$ is the satellite of $M$, and the orbit is a gravitational bound
    2. The system is isolated from other masses. This allows us to neglect any small effect due to outside masses
 - Kepler’s Laws of Planetary Motion
    - First Law – The orbit of each planet about the Sun is an ellipse with Sun at the focus
    - Second Law – Each planet moves so that imaginary line drawn from the sun sweeps out areas in equal times
    - Third Law – The ratio of the squares of the periods of any 2 planets about the Sun is equal to the ratio of the cubes of their average distances from the sun, $\frac{T_1^2}{T_2^2} = \frac{r_1^3}{r_2^3}$
 - Other important Formula
    - $T^2 = \frac{4\pi r^3}{GM}$

# work, energy, and energy resources


### work: the scientific definition
 - Work is done when there is a transfer of energy
 - Alternatively, a force must be exerted and there must be a displacement in the direction of the force
 - For one dimensional work, $W=Fd\cos\theta$
 - The units of work are, $J$, or joules or Newton meters

### kinetic energy and the work-energy theorem
 - Where there is net force, there is net work
 - When force changes (not constant) most of the time $W=Fd\cos\theta$ will not work
 - When speed changes we can derive the formula $W_{net}= \frac{mv_2^2}{2} – \frac{mv_1^2}{2}$
 - This allows us to see that kinetic energy or, $KE = \frac{mv^2}{2}$

### gravitational potential energy
 - Gravitational Potential Energy is also known as work done against gravity
 - When you do work in the y-axis you are either working against or with gravity
 - This is calculated by $W = mgh$
 - $PE = mgh$

### conservative forces and potential energy
 - Potential Energy and Conservative Forces
    - Work is done by forces an some forces have special characteristics
    - A conservative force is one
        - This is where only the starting and ending points of motion matter and the path taken isn’t considered
    - Potential energy – the energy a system has due to positon, shape, or configuration. It is stored energy that is completely recoverable
 - Potential Energy of a Spring
    - For potential energy in a spring, we use Hooke’s Law
    - $W = \frac{kx^2}{2}$ where $k$ is the springs constant, and $x$ is the displacement of its original position
 - Conservation of Mechanical Energy 
    - When dealing with conservative forces, energy isn’t lost
    - Therefore, the initial work and energy should equal the final
    - $W_2 = W_1$
    - $\frac{mv_2^2}{2} + mgh_2 = \frac{mv_1^2}{2} + mgh_1$
    - $KE_2 + PE_2 = KE_1 + PE_1$
    - When kinetic and potential energy is added, this is known as mechanical energy

### non-conservative forces
 - Non-conservative Forces and Friction
    - Forces are either conservative or non-conservative
    - A non-conservative force is one where work depends on the path taken
    - Friction is a non-conservative force
    - Because of its dependence on the path taken, there is no potential energy associated with it
    - Non-conservative forces either add or remove mechanical energy to the system 
    - Ex- energy lost as heat
 - How the Work-Energy Theorem Applies
    - Since kinetic energy is equal to the net work we can say that $W_{net}=\Delta KE$
    - But also since the net work is equal to the sum of conservative and non-conservative forces we can say $W_{nc} + W_c = \Delta KE$
    - It is also known that the work done by conservative forces is the loss of potential energy, -ΔPE
    - This gives us $W_{nc} = \Delta PE + \Delta KE$
    - Finally, the final and initial aspects can be added

### conservative of energy
 - ‘Total energy is constant in any process. It may change in form or be transferred from one system to another, but the total remains the same.’
 - Other Forms of Energy than Mechanical Energy
    - We consider all other forms of energy in one grouping because of our level of physics
    - We denote other energy by, $OE$
    - This gives us $KE_1 + PE_1 + W_{nc} + OE_1 = KE_2 + PE_2 + OE_2$
 - Some Other Forms of Energy
    - Some other forms of energy are electrical, chemical, radiant, nuclear, or thermal energy
 - Efficiency 
    - Efficiency is the output of useful work
    - Its calculated by $Eff = \frac{Work Output}{Total energy input} = \frac{W_{out}}{E_{in}}$
    
### power
 - Power is defined as the rate at which work is done
 - Power is calculated by, $P = \frac{W}{t}$
 - The SI units for power is $\frac{J}{s}$ or watts
 - 1 MW is 1,000,000 W
 - Power and Energy Consumption
    - Energy companies measure energy in kW*h
    - This can be found through the energy equation
    - $P=\frac{W}{t} \rightarrow Pt=W \rightarrow Pt=E$

# linear momentum and collisions


### linear momentum and force 
 - Linear Momentum
    - Defined as the product of a systems mass multiplied by the velocity
    - Expressed as $p=mv$, where $p$ is known as momentum
 - Momentum and Newton’s Second Law
    - Momentum was once called the ‘quantity of motion’
    - The net force equals the change in momentum of a system divided by the time over which it changes
     - $F_{net}= \frac{\Delta p}{\Delta t}$, where $p$ is the change in momentum, and $t$ is the change in time

### impluse
 - The effect of a force depends on the magnitude of the force and how long the force acts
 - Impulse is written $\Delta p = F_{net}t$
 - Impulse can also be written as $J = mv_2 - mv_1$

### conservation of momentum
 - Momentum is an important quantity because it is conserved
 - This means that $\Delta p$ must be constant
 - $\Delta p_1 + \Delta p_2 = 0$
 - If this were wrote in the conservation of momentum form, it would look like $P_{tot} = P_{tot}’$
 - Isolated system – defined to be one for which the net external force is zero

### elastic collisions in one dimension 
 - Consider 2 object moving along a line and they collide
 - When they collide and separate, they create an elastic collision
 - An elastic collision is where internal kinetic energy is conserved
 - Internal kinetic energy is the sum of the kinetic energies of the objects system
 - $\Delta p_2 = \Delta p_1$ and $\Delta KE_2 = \Delta KE_1$
 - Known as the ‘hit and split’ collision
 - There are shortcut formulas for the final velocities

### inelastic collision in one dimension
 - In elastic collisions, energy is conserved
 - In inelastic collisions, energy is not conserved
 - Inelastic collision – a collision in which internal kinetic energy changes (not conserved)
 - Known as the ‘hit an stick’ collisions
 - The conservation of energy formula cannot be applied here

### collisions of point masses in two dimensions
 - When dealing with a problem in 2 dimensions, $x$ and $y$ components are bought out
 - Formulas for the momentum in the $x$ and $y$ direction are derived

# statics and torque


### the first condition for equilibrium
 - The first condition to achieve equilibrium is the net force must be 0
 - This means that the net force must equal zero in the $x$ and $y$ directions
 - There are 2 types of equilibrium, static (motionless) and dynamic (constant velocity)

### the second condition for equilibrium
 - Torque – the second condition to achieve equilibrium involves avoiding accelerated rotation (0 or constant angular velocity)
 - Torque is defined mathematical as $\tau = rF\sin\theta$, where $\tau$ is torque, $r$ is the distance from the pivot point, $F$ is the magnitude of the force, $\theta$ is the angle between the force and the vector
 - $\Delta \tau = \tau_1 + \tau_2 …$

### stability
 - There are 3 types of equilibrium: stable, unstable, and neutral
    - Stable Equilibrium – when displaced from equilibrium, it experiences a net force or torque in a direction opposite to its displacement
    - Unstable Equilibrium – When displaced from equilibrium, it experiences a net force or torque in the same direction as it was displaced in
    - Neutral Equilibrium – Is where its equilibrium is independent of its displacement from its original position

### applications of statics, including problem-solving strategies
 1. the first step is to determine whether or not the system is in static equilibrium. This condition is always the case when the acceleration of the system is zero and accelerated rotation does not occur
 2. Draw an FBD
 3. Solve the problem by applying either or both of the conditions for equilibrium ($F_{net}=0$ and $\tau = 0$)
 4. Check to see if the answer you achieve is reasonable

# rotational motion and angular momentum


### angular acceleration
 - Uniform circular motion and gravitation used only uniform motion (constant speed)
 - The equation for angular velocity is defined as $\omega=\frac{\theta}{t}$
 - The relation between angular velocity and linear velocity is $v = r\omega$
 - When there is no constant velocity the object is in acceleration 
 - Angular acceleration looks like $\alpha = \frac{\Delta \omega}{t}$ where the units are $\frac{rads}{s^2}$
 - If $\omega$ is increasing, $\alpha$ is positive; if $\omega$ is decreasing, $\alpha$ is negative
 - There is also tangential acceleration present
 - Linear or tangential acceleration is the change in velocities magnitude represented by the variable at
 - Centripetal acceleration is the change in the velocities direction, $a_c$
 - $a_t = r\alpha$

### kinematics of rotational motion
 - The kinematics of rotational motion describes relationship between the rotation angle, angular velocity, angular acceleration and time
 - Using the equations from kinematics, we can translate them to rotational motion
 - | Rotational | Translational | - |
   | --- | ---- | --- |
   | $\theta = \omega t$ | $d=vt$ |  |
   | $\omega_2 = \omega_1 + \alpha t$ | $v_2= v_1 + at$ | Constant $a/\alpha$ |
   | $\theta = \omega_1 t + \frac{\alpha t^2}{2}$ | $d = v_1t + \frac{at^2}{2}$ | Constant $a/\alpha$ |
   | $\omega_2^2 = \omega_1^2 + 2\alpha\theta$ | $v_2^2 = v_1^2 + 2ad$ | Constant $a/\alpha$ |

### fynamics of rotational motion: rotational inertia 
 - Using Newton’s 2nd law, $F_{net} = ma$, and the expression $a = r\alpha$, we can see that $F_{net} = mr\alpha$
 - Since we know torque is $rF$, if we multiply both sides by $r$, we get $rF = mr^2\alpha$ or in other terms $\tau = I\alpha$ where $I$ is the moment of inertia 
 - Rotational Inertia and Moment of Inertia
    - $I$ is the sum of all $mr^2$ for all point masses which it is composed of
    - $I = \Sigma mr^2$

### rotational kinetic energy: work and energy revisited
 - Work must be done to rotate objects
 - This can be written as $W_{net} = Fs$
 - $W_{net}$ can also be written as $I\alpha\theta$
    - There is also rotational kinetic energy which is defined as $KE_{rot} = \frac{I\omega^2}{2}$
    - $W_{net} =\frac{I\omega_2^2}{2} - \frac{I\omega_1^2}{2}$  

### angular momentum and its conservation 
 - Every rotational momentum has a direct translational analog
 - Angular momentum, $L$, is defined as $L = I\omega$ where the units are $kg \frac{m^2}{s}$
 - There is also a relationship between torque and angular momentum
 - $\tau_{net} = \frac{\Delta L}{t}$
 - The conservation of momentum principles can be applied here
