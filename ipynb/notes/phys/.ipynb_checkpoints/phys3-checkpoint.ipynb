# elasticity: stress and strain

### stretching, compressing, and Young’s modulus
 - When an object under goes some change in shape by a force, this is known as deformation
 - A change in length, $\Delta L_0$, is made when a force is applied parallel to its length, $L_0$, either stretching it or compressing it
 - $F = \frac{Y\Delta LA}{L}$, where $Y$ is the young’s modulus ($\frac{N}{m^2}$), $\Delta L$ is the length stretched, $L$ is the original length, and $A$ is the surface area
 - The surface area is the area in which the force is distributed on
 
### shear deformation and the shear modulus 
 - When there is a deformation, $\Delta x$, and it is perpendicular to, $L_0$, there is a sideways stress or shear force
 - $F = \frac{S\Delta xA}{L}$, where $S$ is the shear modulus, $\Delta x$ is the change length
 
### volume deformation and the bulk modulus
 - When there is a change in volume or a deformation in the volume, there is compression in all directions equally
 - To compress an object at the molecular level you need to force the molecules closer together
 - $F = \frac{B\Delta VA}{V}$, where $B$ is the bulk modulus, $\Delta V$ is the change in volume, and $V$ is the original volume
 
### general
 - In general, $\frac{F}{A}$ is called the stress of an object and $\frac{\Delta V}{V}$, $\frac{\Delta L}{L}$, $\frac{\Delta x}{L}$ are the strains
 - Stress $= \frac{N}{m^2}$, strain = unitless , moduli = $\frac{N}{m^2}$
 - $1 \frac{N}{m^2} = 1 Pa$
 - | Strain | Moduli | = Stress |
   | --- | --- | --- |
   | $\frac{\Delta L}{L}$ | $Y$ | $=\frac{F}{A}$ |
   | $\frac{\Delta x}{L}$ | $S$ | $=\frac{F}{A}$ |
   | $\frac{\Delta V}{V}$ | $B$ | $=\frac{F}{A}$ |

# energy

### work and energy
 - One Dimensional Work: $W = Fd\cos\theta$ ($J$)
 - Kinetic Energy: $KE = \frac{mv^2}{2}$
 - Gravitational Potential Energy: $U = mgh$
 - Spring Potential Energy: $U = \frac{kx^2}{2}$ 
 - Conservation of Energy: $KE_1 + U_1 = KE_2 + U_2$ 
 - Non-Conservative Work: $W_{nc} + W_c = \Delta KE$
 - Efficiency $= \frac{Work \ Output}{Total \ Energy \ Input} = \frac{W_{out}}{E_{in}}$ 
 - Power $= \frac{Work}{Time}$ ($\frac{J}{s}$ or Watts)

# fluid statics

### what is fluid?
 - Most common states of matter are solid, liquid, and gas
 - Solids – have a definite shape, specific volume
 - Liquids – have definite volume, changing shape
 - Gas – have neither a definite volume, or shape

### density
 - Density is the mass per unit volume 
 - $\rho = \frac{m}{V}$, where $\rho$ is density, $m$ is mass, and $V$ is volume, and the units for density are $\frac{kg}{m^3}$
 - Specific gravity
    - This is the ratio of the object density to the density of water
    - $\frac{\rho}{\rho w}$

### pressure
 - Pressure is the force divided by the area perpendicular to the force over which it acts
 - $P = \frac{F}{A}$, where $P$ is the pressure measured in Pascals or $\frac{N}{m^2}$

### variation of pressure with depth in a fluid
 - Pressure change as depth changes
 - $P = \rho hg$, where $\rho$ is the density, $h$ is the depth in the fluid, and $g$ is gravity
 - See notebook for derivation
 - The standard atmospheric pressure or 1 atm is equal to $P_{atm} = 1.01\times 10^5 Pa$

### Pascal’s principle
 - When a fluid is enclosed it is easier to analyze the relation of the pressure, forces, and area
 - Pascals Principle gives a clear relation
 - ‘A change in pressure applied to an enclosed fluid is transmitted undiminished to all portions of the fluid and to the walls of its container’
 - See notebook for application of the principle in hydraulics

### gauge pressure, absolute pressure, and measurement
 - Gauge Pressure – this is the pressure relative to atmospheric pressure
 - Absolute Pressure – the sum of the gauge pressure and the atmospheric pressure
 - $P_{abs} = P_{gp} + P_{atm}$
 - Remember, atmospheric pressure is $1.01\times 10^5 Pa$
 - And since atmospheric pressure and the formula for pressure is known we see
    - $P_2 = P_1 + \rho gh$

### Archimedes principle
 - Buoyant Force – this force is the net upward force on any object in any fluid
 - $F_B = W_{Fluid}  = m_{object}g = \rho Vg$
 - See notebook for derivation
 - If the object is floating, the buoyant force is equal to the magnitude of its weight
 - When an object is submerged, the buoyant force is dependent on the volume, not the weight

# fluid dynamics

### definitions
 - Steady Flow: the velocity of fluid particles at any point is constant as time passes
 - Unsteady Flow: when the velocity changes at any point as time passes
 - Fluids can be compressible or incompressible
 - Fluids can be viscous or non-viscous
 - Ideal fluids are ones that are incompressible and non-viscous
 - When the flow is steady, streamlines are used to represent trajectories

### flow rate and its relation to velocity
 - Flow Rate, $Q$, is defined to be the volume of fluid passing by some location through an area during a period of time
 - $Q = \frac{V}{t}$, where $V$ is volume, $t$ is time, and the units of $Q$ are $\frac{m^3}{s}$
 - If the volume is thought as the area multiplied by the distance, then flow rate can be seen as
    - $Q = Av$
 - The Equation of Continuity
    - $Q_2 = Q_1$
    - $A_2v_2 = A_1v_1$

### Bernoulli’s equation
 - Fluids accelerate toward lower pressure regions
 - According to the pressure-depth relationship, pressure is lower at higher levels and higher at lower levels
 - Bernoulli’s Equation
    - $P_1 + \frac{\rho v_1^2}{2} + \rho gh_1 = P_2 + \frac{\rho v_2^2}{2} + \rho gh_2$
 - See notebook for derivation

### viscosity and laminar flow; Poiseuille’s law 
 - The magnitude of the tangential force required to move a fluid layer at a constant speed is given by
    - $F = \frac{\eta\Delta v}{y}$, where $\eta$ is the coefficient of viscosity, $\Delta v$ is the difference in velocity, and $y$ is the distance between the plates
 - Poiseuille’s Law
    - The volume flow rate rate is given by
    - $Q = \frac{\pi r4(P_2 - P_1)}{8\eta L}$, where $r$ is the radius, $\eta$ is the coefficient of viscosity, and $L$ is the length
 - Reynolds Number
    - Helps find flow patterns
    - $Re = \frac{2v\rho r}{\eta}$, where v is the average velocity through the laming
    - Reynolds Number is unitless
    - Laminar Flow $\rightarrow Re < 2000$
    - Turbulent Flow $\rightarrow Re > 4000$
    - Anything in between is called transition flow

# temperature, kinetic theory, and gas laws

### temperature scales
 - Temperature are commonly reported in degrees Celsius or Fahrenheit (°C or °F)
 - Temperature changes are reported in Celsius degrees or Fahrenheit degrees (C° or F°)
 - The Kelvin Temperature Scale
    - The kelvin scale is the Celsius scale just readjusted to absolute zero (where $P=0$)
    - A one degree temperature change in Celsius is a one degree change in Kelvin
    - $T = T_C + 273.15$
 - Thermometers make use of the change in some physical property with temperature. The physical property that changes is a thermometric property

### linear thermal expansion
 - The relationship of the temperature and the change in the length of the object is written as 
    - $\Delta L = \alpha L_0\Delta T$, where $\Delta L$ is the change in length, $L$ is the original length, $\alpha$ is the coefficient of linear expansion ($C°^{-1}$), $\Delta T$ is the temperature change in centigrade

### volume thermal expansion	
 - The relationship of the temperature and the change in the volume of the object is
    - $\Delta V = \beta V_0\Delta T$, where $\Delta V$ is the change in volume, $V$ is the original volume, $\beta$ is the coefficient of volume expansion ($C°^{-1}$)

### equilibrium between phases of matter
 - The pressure of vapour that coexists in equilibrium with the liquid is called the equilibrium vapor pressure of the liquid
 - Only when the temperature and vapour pressure correspond to a point on the curved line do the liquid and vapor phases coexist in equilibrium

### humidity
 - Air is a mixture of various gases
 - The total pressure is the sum of partial pressures of the component gases
 - The partial pressures of the water vapour depend on weather conditions
 - It can be low as 0 or high as the vapor pressure of water at the given temperature (rain)
 - Percent Relative to Humidity $= \frac{\partial Pressure \ of \ Water \ Vapour}{Equilibrium \ vapour \ pressure \ of \ water} \times 100$ 
 - 100% relative humidity is called the dew point

### molecular mass, the mole, and Avogadro’s number
 - To compare the mass of atoms, the atomic mass scale was created
 - The unit is called atomic mass units (u)
 - $1 u = 1.66\times 10^{-27} kg$
 - One mole of a substance contains as many particles as there are in 12 grams of the isotope carbon-12
 - The number of atoms per mole is Avogadro’s Number ($N_A = 6.022\times 10^{23} mol^{-1}$)
 - $n = \frac{N}{N_A}$, where $n$ is the number of moles, $N$ is the number of atoms

### the ideal gas law
 - An ideal gas is an idealized model for real gases that have sufficiently low densities
 - The condition of low density means that the molecules are so far apart that they do not interact except during collisions which one effectively elastic
 - At constant volume, the pressure is inversely proportional to temperature 
 - At constant temperature, pressure is proportional to volume
 - The pressure is also proportional to the amount of gas
 - $PV = nRT$, where $P$ is pressure, $V$ is volume, $n$ is the number moles, $R$ is $8.31 \frac{J}{molK}$, $T$ is temperature
 - Another variation of this equation is 
    - $PV = NkT$, where $N$ is the number of atoms, and $k$ is $1.38\times 10^{-23}$
 - This allows us to see that
    - $\frac{P_f V_f}{T_f}=\frac{P_i V_i}{T_i}$ 

### kinetic theory of gases
 - Particles are in constant, random, motion colliding with each other and the walls of the container
 - Each collision changes the particles speed, thus the atoms and molecules have different speed
 - This gives us a distribution of speeds
 - Force on a single molecule $= F = \frac{2mv}{L}$ 
 - For $N$ molecules average force $= F = \frac{N}{3}\frac{2mv_{rms}}{L}$, where $v_{rms}$ is the root-mean-squared velocity
 - $KE = \frac{mv_{rms}^2}{2} = \frac{3kT}{2}$
 - Internal Energy of a Monoatomic Ideal Gas
    - $U =\frac{3NkT}{2}= \frac{3nRT}{2}$

### diffusion
 - The process in which molecules move from a region of higher concentration to one of a lower concentration is called diffusion
 - Frick’s Law of Diffusion
    - Mass, $m$, of solute that diffuses in a time, $t$, through a solvent contained in a length, $L$, and cross-sectional area, $A$, is 
    - $m = \frac{DA\Delta Ct}{L}$, where $D$ is the diffusion constant, $\Delta C$ is the concentration gradient

# heat and heat transfer models

### heat
 - Heat is defined as the spontaneous flow of energy due to a temperature difference
 - Heat only flows from hot to cold temperature differences
 - Heat is energy
 - Therefore, the temperature can be changed by doing work
        
### temperature and heat capacity 
 - One of the major effects of heat transfer is temperature change
 - Heating (adding energy) increases temperature while cooling (taking out energy) decreases temperature 
 - Transferred energy depends on 3 factors: change in temperature, mass of the system, and the phases of the substance
 - $Q = mc\Delta T$, where $Q$ is the heat energy transfer, and $c$ is the specific heat capacity
        
### phase change and latent heat
 - During a phase change, the temperature of the mixture does not change (provided it is in thermal equilibrium)
 - The energy involved in phase change depends on the number and strength in bond pairs
 - The amount of heat required to phase change an object is found by
     - $Q = mL_F$ or $mL_V$ , where $L_f$ is the latent heat of fusion, and $L_V$ is the latent heat of vaporization
     - The latent heat of fusion is for melting and freezing
     - The latent heat of vaporization is for vaporing and condensing
            
### heat transfer methods
 1. Convection: the process in which heat is carried from one place to another by the bulk movement of a fluid. Convection can be forced
 2. Conduction: the transfer of heat through direct contact. If a substance conducts heat well it is a thermal conductor and if the opposite, then it is a thermal insulator. $Q = \frac{kA\Delta Tt}{L}$
 3. Radiation: the transfer of heat energy through electromagnetic waves. A material that absorbs all EM waves is a blackbody. $Q = e\sigma T^4At$, $e$ is the emissivity, $\sigma$ is $5.678 x 10^{-8}$

# thermodynamics


### thermodynamics
 - The first law of thermodynamics states that the change in internal energy is equal to the net heat transfer into the system minus the net work done by the system
 - $\Delta U = Q – W$
 - If $Q$ is positive than there is a net heat transfer into the system
 - If $W$ is positive then there is net work done by the system
 - So positive $Q$ adds energy while positive $W$ takes energy away and vice versa
 - Note, $\Delta U$ is defined as internal energy

### the first law and simple processes
 - Using heat transfer as a means to get useful work is called a heat engine
 - $W = Q_{in} – Q_{out}$, where $Q_{in}$ is the input heat, $W$ is the useful work done, $Q_{out}$ is the rejected heat
 - $PV$ Diagrams and their Relation to Work 
    - Isobaric: Pressure is held constant. $W = P\Delta V = P(V_f – V_i)$
    - Isochoric: volume is held constant. $\Delta U = Q$ because $W = 0$ 
    - Isothermal: Temperature is held constant. $PV = nRT$, $W = nRT\ln(\frac{V_f}{V_i})$
    - See notebook for their respective graphs
    - Adiabatic Process: A process where there is no heat transfer
    - Reversible Process: A process where both the system and its environment return to the initial states
    - | Name | What is held constant | Work relation |
      | --- | --- | --- |
      | Isobaric | Constant pressure | $W = P\Delta V$ | 
      | Isochoric | Constant volume | $W = 0$ |
      | Isothermal | Constant tempertaure | $Q = W$ |
      | Adiabatic | No heat transfer | $Q = 0$ |

### the second law of thermodynamics and efficiency 
 - The second law of thermodynamics is the heat transfer when it occurs spontaneously from higher to lower temperature bodies but never in the reverse direction
 - The second law can also be described in a different way
    - It is impossible in any system for heat transfer from a reservoir to completely convert to wok in a cyclical process which the system returns to its initial state
 - Heat Engine Efficiency: $Eff = \frac{W}{Q_H}$ or $1 - \frac{Q_C}{Q_H}$ 

### Carnot’s perfect heat engine: the second law revisited
 - A third form of the second law can be written for reversible processes
 - A Carnot engine operating between two given temperatures has the greatest possible efficiency at any engine operating between these two temperatures
 - $Eff = 1-\frac{T_c}{T_h}$


### applications of thermodynamics: heat pumps/refrigerators
 - Heat Pumps
    - Coefficient of Performance: $COP_{heating} = \frac{|Q_H|}{W} = \frac{Q_C + W}{W} = COP_{cooling} + 1$ 
    - $|W| = |Q_H| (1 - \frac{T_C}{T_H})$
 - Refrigeration
    - Coefficient of Performance: $COP_{cooling} = \frac{|Q_C|}{W} = \frac{Q_C}{W}$ 
    - $|W| = |Q_C| (1 -\frac{T_C}{T_H})$

### entropy and the second law of thermodynamics
 - When analyzing the Carnot engine, we observed that the ratio of the hot and cold temperatures is equal to the ratio of the heat energies
 - After rearranging you achieve the ratio of heat energy over temperature
 - This is the definition of entropy
 - $\Delta S = \frac{Q}{T}$
 - $\Delta S = \Delta S_H - \Delta S_C$
 - $\Delta S = nC\ln(\frac{T_f}{T_i})$
 - $S = k\ln(W)$, where $W$ is $\frac{N!}{n_1! n_2!}$

# oscillatory motion and waves

### Hooke’s haw revisited
 - $F = -kx$
 - Energy in Hooke’s Law: $U = \frac{kx^2}{2}$

### period and frequency oscillations
 - Period is the amount of time needed to complete one oscillation
 - Frequency is the amount of cycles per second
 - $f = \frac{1}{T}$

### simple harmonic motion
 - SHM allows us to relate the period and frequency to the spring constant and mass of it
 - $T = 2\pi\sqrt{\frac{m}{K}}$
 - $f = \frac{1}{2\pi}\sqrt{\frac{K}{m}}$
 - The position of the SHM can be found here
    - $x = A\cos\theta = A\cos(\omega t)$, where $A$ is the amplitude and $\omega$ is the angular frequency
 - The velocity can be found in
    - $v = -v\sin\theta = -A\omega\sin(\omega t)$, when $v = -A\omega$, that is max velocity
 - The acceleration can be found
    - $a = -A\omega^2\cos(\omega t)$

### simple pendulums
 - We can use the angle approximation in this situation and can see that $\theta = \frac{s}{L}$
 - $\omega = \sqrt{\frac{g}{L}}$
 - $T = 2\pi \sqrt{\frac{L}{g}}$

### energy and the simple harmonic oscillators
 - $E_{Tot} = KE + PE_{elastic}$
 - $\frac{mv_1^2}{2} + \frac{kx_1^2}{2} = \frac{mv_2^2}{2} + \frac{kx_2^2}{2}$

### forced oscillations and resonance
 - Natural Frequency: the frequency at which a system would oscillate if there were no driving or damping forces
 - The phenomenon of driving a system with a frequency equal to its natural frequency
 - A system being driven at its natural frequency is said to resonate

### waves
 - A wave is a disturbance that propagates from the place it was created
 - The wave velocity is defined as the speed at which the disturbance moves
 - $v_w = \frac{\lambda}{T} = f\lambda$
 - Waves can also be transverse or longitudinal
 - Transverse waves oscillate in positions perpendicular to the velocity
 - Longitudinal waves oscillate parallel to the velocity
 - Speed of waves on a string
    - $v = \sqrt{\frac{FL}{m}}$  

### superposition and interference
 - When 2 or more waves interfere, they come together and create a phenomenon called superposition
 - When waves line up crest to crest or trough to trough, they create constructive interference
 - When waves line up crest to trough or vice versa, they create a pure destructive interference
 - When a wave seems to alternate between constructive and destructive interference, they create a standing wave
 - When the wavelength of a wave is fixed, the fundamental frequency is $\lambda = 2L$
    - $f = \frac{v}{\lambda} = \frac{v_w}{2L}$
 - Beat Frequency is the wave resulting from the superposition of the 2 similar frequency waves
    - $f_B = |f_1 – f_2|$

# physics of hearing

### sound
 - Hearing is the human perception of sound
 - The physical phenomenon of sound is known as pressure waves
 - The eardrum receives these pressure waves and perceives them as we do
 
### speed of sound, frequency, and wavelength
 - Our perception of frequency is known as pitch
 - Speed of sound $= v_w = f\lambda$
 - The speed of sound varies with medium, in air it is about $331 \frac{m}{s}$
 - The temperature also affects the speed of sound
    - $v_w = (331)\sqrt{\frac{T}{273}}$

### sound intensity and sound level
 - We know intensity is $\frac{Power}{Area}$
 - Sound intensity level can be found in the equation $\beta (dB) = 10\log_{10}(\frac{I}{I_0})$
 - $I_0$ is the threshold of hearing and equates to $1\times 10^{-12}\frac{W}{m^2}$

### the doppler effect
 - The Scientific definition of the Doppler effect is an increase or decrease in the frequency of sound/light as the source and observer move toward or away from each other
 - Stationary Observer
    - Towards observer: $f_0 = f_S(\frac{1}{1-\frac{v_s}{v}})$
    - Away from observer: $f_0 = f_S(\frac{1}{1+\frac{v_s}{v}})$
 - Moving Observer
    - Towards source: $f_0 = f_S (1 + \frac{v_0}{v})$
    - Away from source: $f_0 = f_S (1 - \frac{v_0}{v})$
 - General Case
    - $f_0 = f_S (\frac{1 \pm \frac{v_s}{v}}{1 \pm \frac{v_0}{v}})$

### sound interference and resonance
 - Interference: a fundamental aspect of waves that observing interference is proof that something is a wave
 - Two waves in phase create constructive interference
 - Two waves out of phase create destructive interference
 - Diffraction: the bending of a wave around an obstacle or the edges of an opening is called diffraction
 - Beats: two overlapping waves with slightly different frequencies give rise to the phenomena of beats. 
 - String fixed at both ends
    - $f_n = n(\frac{v}{2L})$
 - Tube open at both ends
    - $f_n = n(\frac{v}{2L})$
 - Tube open at one end
    - $f_n = n(\frac{v}{4L})$

# geometric optics, vision, and optical instruments

### the ray aspect of light
 - A ray is a straight line that originates at some point
 - Geometric Optics is the part of optics dealing with the ray aspect of light is called geometric optics

### the law of reflection
 - The law of reflection states that the incident angle is equal to the angle of reflection
 - $\theta_i = \theta_R$

### the law of refraction
 - Refraction: the changing of a lights rays direction (loosely called bending) when it asses through variation in matter 
 - The speed of light $= c \approx 3\times 10^8 \frac{m}{s}$
 - The speed of light changes through mediums
 - $n = \frac{c}{v}$, where $n$ is the index of refraction, $c$ is the speed of light, and $v$ is the speed of light in the medium
 - Law of Refraction (Snell’s Law)
    - $n_1\sin\theta_1 = n_2\sin\theta_2$

### total internal reflection
 - Total internal reflection is when all of the light is reflected from the surface
 - This happens at the critical angle
 - Critical angle is the incident angle that produces an angle of refraction of 90° 
 - $\theta_C = \sin^{-1}(\frac{n_2}{n_1})$

### dispersion: the rainbow and prisms
 - Dispersion is the spreading of white light into its full spectrum of wavelengths
 - This happens because different wavelengths refract differently than others

### image formation by lenses
 - Converging/Convex Lenses
    - This lens is one where the light ways that enter parallel to its axis cross one another at a single point on the opposite side
 - Focal Point is the point at which light rays cross
 - Focal Length is the distance from the centre of the lens to the focal point
 - Power of a lens if inversely proportional to the focal length of a lens
    - $D = \frac{1}{f}$ where the units are diopters
 - Diverging Lens
    - This lens is one where the light rays bend away from the parallel axis
 - Magnification $= m = \frac{h_i}{h_o} = \frac{-d_i}{d_o}$, where $h_i$ is image height, $h_o$ is object height, $d_i$ is image distace, and $d_o$ is object distance
 - $\frac{1}{f} = \frac{1}{d_o} + \frac{1}{d_i}$
 - Summary of sign conventions for lenses
     - $f$ is $+$ for a converging lens
     - $f$ is $-$ for a diverging lens
     - $d_o$ is $+$ if the object is to the left of the lens
     - $d_o$ is $-$ if the object is to the right of the lens
     - $d_i$ is $+$ if the image is formed to the right of the lens (real image)
     - $d_i$ is $-$ if the image is formed to the left of the lens (virtual image)
     - $m$ is $+$ for an upright image
     - $m$ is $-$ for an inverted image

# wave optics

### linear superposition
 - When 2 or more waves pass through a point, they create a phenomenon called superposition
 - 2 waves in phase create constructive interference
 - 2 waves out of phase create destructive interference
 - If constructive or destructive interference is continually occurring at a point the sources must be coherent

### Young’s double slit experiment 
 - In Young’s experiment, he displayed that light acts like a wave
    - Bright fringe: $\sin\theta = \frac{m\lambda}{d}$
    - Dark fringe: $\sin\theta = \frac{(m+\frac{1}{2})\lambda}{d}$ 

### thin film interference
 - Because of reflection and refraction, light waves split up when passing through a thin film
 - $\lambda_{film} = \frac{\lambda}{n}$

### diffraction
 - Diffraction is the bending of waves around obstacles or an edge of an opening
 - Huygens’ Principle
    - Every point on a wave front acts as a source of tiny wavelets that move forward with the same speed as the wave; the wave front at a later instant is the surface that is tangent to the wavelets.
 - Dark Fringes: $\sin\theta = \frac{m\lambda}{a} $
 - Circular Diffraction: $\sin\theta = \frac{1.22\lambda}{D}$ 
 - Diffraction Grating is where there is an arrangement of closely spaced parallel slits
    - Principle Maxima of Diffraction Grating: $\sin\theta = \frac{m\lambda}{d} $
